```python

#criar função de cadastrar usuaŕio e cadastrar conta bancária


#criar usuário cliente:
#o programa deve armazenar os usuários em uma lista, usuário tem nome, data de nascimento, cpf, e endereço.
#o endereço é uma string com o formato: logradouro, numero - bairro - cidade/sigla estado
#deve ser armazenado somente os números do cpf
#não pode ter dois usuários com o mesmo cpf


#criar conta corrente:
#o programa deve armazenar contas em uma lista, uma conta é composta por agencia, número de conta e usuário
#O número de conta é sequencial, iniciando em 1.
#o número de agencia é fixo: 0001
#O usuário pode ter mais de uma conta, mas uma conta pertence somente a um usuário.

#dica
#para vincular um usuário a uma conta, filtre a lista de usuários buscando o número do cpf informado para cada usuário da lista.



```

```python
usuarios = []
contas_correntes = []
proximo_numero_conta = 1

def criar_usuario(nome, data_nascimento, cpf, endereco):
    for usuario in usuarios:
        if usuario['cpf'] == cpf:
            print(f"Já existe um usuário cadastrado com o CPF {cpf}.")
            return None
    
    usuario = {
        'nome': nome,
        'data_nascimento': data_nascimento,
        'cpf': cpf,
        'endereco': endereco
    }
    usuarios.append(usuario)
    print(f"Usuário {nome} cadastrado com sucesso.")
    return usuario

def criar_conta(usuario):
    global proximo_numero_conta
    
    for conta in contas_correntes:
        if conta['usuario']['cpf'] == usuario['cpf']:
            print(f"O usuário {usuario['nome']} já possui uma conta.")
            return None
    
    conta = {
        'agencia': '0001',
        'numero_conta': proximo_numero_conta,
        'usuario': usuario
    }
    contas_correntes.append(conta)
    proximo_numero_conta += 1
    print(f"Conta corrente criada para o usuário {usuario['nome']} - Agência: {conta['agencia']} - Conta: {conta['numero_conta']}.")
    return conta

def extrato_conta(saldo, /, *, extrato):
    if not extrato:
        return "Não foram realizadas movimentações.", saldo
    else:
        extrato_completo = f"\n================ EXTRATO ================\n{extrato}\nSaldo: R$ {saldo:.2f}\n=========================================="
        return extrato_completo, saldo

def deposito(saldo, valor, extrato, /):
    if valor <= 0:
        return '\nValor inválido.', saldo, extrato
    else:
        saldo += valor
        extrato += f"Depósito: R$ {valor:.2f}\n"
        return f'\nDepósito realizado. Seu saldo é: R$ {saldo:.2f}', saldo, extrato

def saque(*, valor, limite, saldo, extrato, limite_saques, numero_saques):
    if numero_saques >= limite_saques:
        return '\nLimite de saques diário atingido.', saldo, extrato, numero_saques
    else:
        if valor <= 0:
            return '\nValor inválido.', saldo, extrato, numero_saques
        elif valor > limite:
            return '\nO valor máximo de saque é 500.', saldo, extrato, numero_saques
        elif valor > saldo:
            return f'\nVocê não tem fundos suficientes na sua conta: \nValor de Saque: R$ {valor:.2f} \nSaldo: R$ {saldo:.2f}', saldo, extrato, numero_saques
        else:
            saldo -= valor
            extrato += f"Saque: R$ {valor:.2f}\n"
            numero_saques += 1
            return f'\nSaque realizado. Seu saldo é: R$ {saldo:.2f}', saldo, extrato, numero_saques

def login():
    cpf = input("Digite seu CPF (somente números): ")
    numero_conta = int(input("Digite o número da sua conta: "))
    
    usuario_encontrado = None
    for usuario in usuarios:
        if usuario['cpf'] == cpf:
            usuario_encontrado = usuario
            break
    
    if not usuario_encontrado:
        print(f"Usuário com CPF {cpf} não encontrado.")
        return None, None
    
    conta_encontrada = None
    for conta in contas_correntes:
        if conta['usuario']['cpf'] == cpf and conta['numero_conta'] == numero_conta:
            conta_encontrada = conta
            break
    
    if not conta_encontrada:
        print(f"A conta {numero_conta} não está vinculada ao usuário com CPF {cpf}.")
        return None, None
    
    print(f"Login efetuado com sucesso para {usuario_encontrado['nome']} - Agência: {conta_encontrada['agencia']} - Conta: {conta_encontrada['numero_conta']}.")
    return usuario_encontrado, conta_encontrada

def escolher_login_ou_criar_conta():
    while True:
        opcao = input("Digite 'login' para entrar na sua conta ou 'criar' para criar uma nova conta: ").strip().lower()
        
        if opcao == 'login':
            return 'login'
        elif opcao == 'criar':
            return 'criar'
        else:
            print("Opção inválida. Por favor, digite 'login' ou 'criar'.")

opcao = escolher_login_ou_criar_conta()

if opcao == 'criar':
    nome = input("Digite seu nome: ")
    data_nascimento = input("Digite sua data de nascimento (DD/MM/AAAA): ")
    cpf = input("Digite seu CPF (somente números): ")
    endereco = input("Digite seu endereço no formato 'logradouro, numero - bairro - cidade/estado': ")
    
    usuario = criar_usuario(nome, data_nascimento, cpf, endereco)
    if usuario:
        criar_conta(usuario)
    else:
        print("Não foi possível criar o usuário ou a conta.")

usuario_logado = None
if opcao == 'login':
    while not usuario_logado:
        usuario_logado, conta_logada = login()

saldo = 0
limite = 500
extrato = ""
numero_saques = 0
LIMITE_SAQUES = 3

while True:
    menuValidated = False
    while not menuValidated:
        menu = input('\nEntre com a opção desejada: \n [d] - Depositar \n [s] - Sacar \n [e] - Extrato \n [q] - sair\n')
        
        if menu not in ['d', 's', 'e', 'q']:
            print('\nPor gentileza, entre com uma das opções disponíveis.')
        else:
            menuValidated = True

    if menu == 'd':
        valor = input('\nInsira o valor do depósito: ')
        try:
            valor = float(valor)
        except ValueError:
            print("\nO depósito só aceita valores numéricos.")
            continue

        mensagem, saldo, extrato = deposito(saldo, valor, extrato)
        print(mensagem)
        
    elif menu == 's':
        valor = input('\nInsira o valor do saque: ')
        try:
            valor = float(valor)
        except ValueError:
            print("\nO saque só aceita valores numéricos.")
            continue

        print(f'\nChegou aqui - valor: {valor}')

        mensagem, saldo, extrato, numero_saques = saque(valor=valor, limite=limite, saldo=saldo, extrato=extrato, limite_saques=LIMITE_SAQUES, numero_saques=numero_saques)
        print(mensagem)

    elif menu == 'e':
        mensagem, saldo = extrato_conta(saldo, extrato=extrato)
        print(mensagem)

    elif menu == 'q':
        break

print("\nOperações encerradas. Obrigado por utilizar nossos serviços!")
```
